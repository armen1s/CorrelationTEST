```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA


data = load_iris()
X = data.data  # Признаки
y = data.target  # Метки классов
target_names = data.target_names
```

```python
# Применяем PCA для снижения размерности до 2
pca = PCA(n_components=2)
X_r = pca.fit_transform(X)

print(f"Доля объяснённой дисперсии по каждой компоненте: {pca.explained_variance_ratio_}")
```

```python

colors = ['navy', 'turquoise', 'darkorange']
plt.figure(figsize=(8, 6))

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], alpha=0.8, color=color, label=target_name)

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
plt.xlabel('Первая главная компонента')
plt.ylabel('Вторая главная компонента')
plt.grid()
plt.show()
```
