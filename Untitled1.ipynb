```python
!pip install yfinance -q

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torch.nn.utils.parametrizations import weight_norm
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)

print("ðŸ“¥ Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° Ñ€ÐµÐ°Ð»ÑŒÐ½Ñ‹Ñ… Ð´Ð°Ð½Ð½Ñ‹Ñ… BTC-USD (Ð¿Ð¾ÑÐ»ÐµÐ´Ð½Ð¸Ðµ 700 Ð´Ð½ÐµÐ¹)...")
# ÐžÐ³Ñ€Ð°Ð½Ð¸Ñ‡Ð¸Ð²Ð°ÐµÐ¼ Ð¿ÐµÑ€Ð¸Ð¾Ð´, Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ñ‚ÑŒ Ñ‡Ð°ÑÐ¾Ð²Ñ‹Ðµ ÑÐ²ÐµÑ‡Ð¸
df = yf.download("BTC-USD", period="700d", interval="1h", auto_adjust=True)

if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

if len(df) == 0:
    print("âŒ Ð”Ð°Ð½Ð½Ñ‹Ðµ Ð½Ðµ Ð·Ð°Ð³Ñ€ÑƒÐ¶ÐµÐ½Ñ‹. ÐŸÐ¾Ð¿Ñ€Ð¾Ð±ÑƒÐ¹Ñ‚Ðµ ÑƒÐ¼ÐµÐ½ÑŒÑˆÐ¸Ñ‚ÑŒ Ð¿ÐµÑ€Ð¸Ð¾Ð´ Ð¸Ð»Ð¸ Ð¿Ñ€Ð¾Ð²ÐµÑ€Ð¸Ñ‚ÑŒ Ð¸Ð½Ñ‚ÐµÑ€Ð½ÐµÑ‚.")
else:
    print(f"âœ… Ð£ÑÐ¿ÐµÑˆÐ½Ð¾ Ð·Ð°Ð³Ñ€ÑƒÐ¶ÐµÐ½Ð¾ {len(df)} ÑÐ²ÐµÑ‡ÐµÐ¹.")
```

```python
def prepare_real_data(df, horizon=12, vol_window=48, threshold=1.2):
    df = df.copy()

    # 1. Ð Ð°ÑÑ‡ÐµÑ‚ Ð²Ð¾Ð»Ð°Ñ‚Ð¸Ð»ÑŒÐ½Ð¾ÑÑ‚Ð¸
    df['returns'] = np.log(df['Close'] / df['Close'].shift(1))
    df['vol'] = df['returns'].rolling(vol_window).std()

    # 2. ÐŸÑ€Ð¸Ð·Ð½Ð°ÐºÐ¸: OHLCV + Ð»Ð¾Ð³Ð°Ñ€Ð¸Ñ„Ð¼ Ð²Ð¾Ð»Ð°Ñ‚Ð¸Ð»ÑŒÐ½Ð¾ÑÑ‚Ð¸ (Ð¼Ð°ÑÑˆÑ‚Ð°Ð±)
    features = df[['Open', 'High', 'Low', 'Close', 'Volume']].copy()
    features = (features - features.rolling(vol_window).mean()) / (features.rolling(vol_window).std() + 1e-9)
    features['log_vol'] = np.log(df['vol'].clip(lower=1e-9))

    # 3. Ð Ð°Ð·Ð¼ÐµÑ‚ÐºÐ° (3 ÐºÐ»Ð°ÑÑÐ°)
    future_ret = np.log(df['Close'].shift(-horizon) / df['Close'])
    labels = np.ones(len(df)) # Class 1: HOLD

    valid = df['vol'] > 0
    labels[valid & (future_ret > df['vol'] * threshold)] = 2 # Class 2: BUY
    labels[valid & (future_ret < -df['vol'] * threshold)] = 0 # Class 0: SELL

    # 4. ÐžÑ‡Ð¸ÑÑ‚ÐºÐ°
    df_final = features.copy()
    df_final['target'] = labels
    df_final = df_final.dropna()

    return df_final.drop(columns=['target']).values.astype(np.float32), df_final['target'].values.astype(np.int64)

class TradingDataset(Dataset):
    def __init__(self, X, y, seq_len=60):
        self.X, self.y, self.seq_len = X, y, seq_len
    def __len__(self):
        return len(self.X) - self.seq_len
    def __getitem__(self, idx):
        return self.X[idx : idx + self.seq_len].T, self.y[idx + self.seq_len - 1]

X_data, y_data = prepare_real_data(df)
full_dataset = TradingDataset(X_data, y_data)

train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_ds, test_ds = torch.utils.data.random_split(full_dataset, [train_size, test_size])

train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
test_loader = DataLoader(test_ds, batch_size=64, shuffle=False)

print(f"ðŸ“Š Ð Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ ÐºÐ»Ð°ÑÑÐ¾Ð²: {np.bincount(y_data)}")
```

```python
class CausalConv1d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size, dilation):
        super().__init__()
        self.pad = (kernel_size - 1) * dilation
        self.conv = weight_norm(nn.Conv1d(in_channels, out_channels, kernel_size, padding=0, dilation=dilation))
    def forward(self, x):
        return self.conv(F.pad(x, (self.pad, 0)))

class TemporalBlock(nn.Module):
    def __init__(self, n_inputs, n_outputs, kernel_size, dilation, dropout=0.2):
        super().__init__()
        self.net = nn.Sequential(
            CausalConv1d(n_inputs, n_outputs, kernel_size, dilation),
            nn.ReLU(), nn.Dropout(dropout),
            CausalConv1d(n_outputs, n_outputs, kernel_size, dilation),
            nn.ReLU(), nn.Dropout(dropout)
        )
        self.downsample = nn.Conv1d(n_inputs, n_outputs, 1) if n_inputs != n_outputs else None
        self.relu = nn.ReLU()
    def forward(self, x):
        out = self.net(x)
        res = x if self.downsample is None else self.downsample(x)
        return self.relu(out + res)

class TradingTCN(nn.Module):
    def __init__(self, num_inputs):
        super().__init__()
        num_channels = [32, 32, 32, 32]
        layers = []
        for i in range(len(num_channels)):
            d = 2**i
            layers.append(TemporalBlock(num_inputs if i==0 else num_channels[i-1], num_channels[i], 3, d))
        self.tcn = nn.Sequential(*layers)
        self.fc = nn.Linear(num_channels[-1], 3)
    def forward(self, x):
        return self.fc(self.tcn(x)[:, :, -1])

model = TradingTCN(num_inputs=6).to(device)
```

```python
counts = np.bincount(y_data)
weights = torch.tensor(1.0 / counts, dtype=torch.float32).to(device)
weights = weights / weights.sum() * 3.0

criterion = nn.CrossEntropyLoss(weight=weights)
optimizer = optim.AdamW(model.parameters(), lr=0.001)

def train(epochs=25):
    for epoch in range(1, epochs + 1):
        model.train()
        l_sum = 0
        for X_b, y_b in train_loader:
            X_b, y_b = X_b.to(device), y_b.to(device)
            optimizer.zero_grad()
            loss = criterion(model(X_b), y_b)
            loss.backward()
            optimizer.step()
            l_sum += loss.item()

        model.eval()
        correct = 0
        with torch.no_grad():
            for X_b, y_b in test_loader:
                X_b, y_b = X_b.to(device), y_b.to(device)
                correct += (model(X_b).argmax(1) == y_b).sum().item()

        print(f"Epoch {epoch:02d} | Loss: {l_sum/len(train_loader):.4f} | Acc: {100*correct/len(test_ds):.2f}%")

train(25)
```

```python
from sklearn.manifold import TSNE

def plot_real_results(num_points=300):
    model.eval()
    lats, preds, probs, prices = [], [], [], []

    # Ð˜Ð½Ð´ÐµÐºÑ Ð½Ð°Ñ‡Ð°Ð»Ð° Ð´Ð»Ñ Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ð¸ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½Ð¸Ñ… num_points
    start_idx = len(full_dataset) - num_points
    if start_idx < 0: start_idx = 0

    print(f"ðŸ“Š ÐŸÐ¾Ð´Ð³Ð¾Ñ‚Ð¾Ð²ÐºÐ° Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ð¸ Ð´Ð»Ñ {num_points} Ñ‚Ð¾Ñ‡ÐµÐº...")

    with torch.no_grad():
        for i in range(start_idx, len(full_dataset)):
            # ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ Ð´Ð°Ð½Ð½Ñ‹Ðµ (ÑÑ‚Ð¾ numpy Ð¼Ð°ÑÑÐ¸Ð²Ñ‹ Ð¸Ð· TradingDataset)
            x_seq, _ = full_dataset[i]

            # Ð˜Ð¡ÐŸÐ ÐÐ’Ð›Ð•ÐÐ˜Ð•: ÐšÐ¾Ð½Ð²ÐµÑ€Ñ‚Ð¸Ñ€ÑƒÐµÐ¼ numpy Ð² torch.tensor Ð¿ÐµÑ€ÐµÐ´ unsqueeze
            x_t = torch.from_numpy(x_seq).float().unsqueeze(0).to(device)

            # ÐŸÑ€Ð¾Ð³Ð¾Ð½ Ñ‡ÐµÑ€ÐµÐ· ÑÐ»Ð¾Ð¸ TCN Ð´Ð»Ñ Ð¿Ð¾Ð»ÑƒÑ‡ÐµÐ½Ð¸Ñ Ð»Ð°Ñ‚ÐµÐ½Ñ‚Ð½Ð¾Ð³Ð¾ Ð¿Ñ€ÐµÐ´ÑÑ‚Ð°Ð²Ð»ÐµÐ½Ð¸Ñ
            lat = model.tcn(x_t)[:, :, -1]
            # ÐŸÑ€Ð¾Ð³Ð¾Ð½ Ñ‡ÐµÑ€ÐµÐ· Ñ„Ð¸Ð½Ð°Ð»ÑŒÐ½Ñ‹Ð¹ ÑÐ»Ð¾Ð¹ Ð´Ð»Ñ ÐºÐ»Ð°ÑÑÐ¸Ñ„Ð¸ÐºÐ°Ñ†Ð¸Ð¸
            out = model.fc(lat)

            lats.append(lat.cpu().numpy()[0])
            preds.append(out.argmax(1).item())
            probs.append(F.softmax(out, dim=1).cpu().numpy()[0])

            # Ð¡Ð¾Ð¿Ð¾ÑÑ‚Ð°Ð²Ð»ÐµÐ½Ð¸Ðµ Ñ Ñ†ÐµÐ½Ð¾Ð¹ Ð¸Ð· Ð¾Ñ€Ð¸Ð³Ð¸Ð½Ð°Ð»ÑŒÐ½Ð¾Ð³Ð¾ DataFrame
            real_idx = len(df) - len(full_dataset) + i
            prices.append(df['Close'].iloc[real_idx])

    # --- Ð’Ð˜Ð—Ð£ÐÐ›Ð˜Ð—ÐÐ¦Ð˜Ð¯ ---
    plt.figure(figsize=(15, 12))

    # Ð“Ñ€Ð°Ñ„Ð¸Ðº 1: Ð¡Ð¸Ð³Ð½Ð°Ð»Ñ‹ Ð½Ð° Ñ†ÐµÐ½Ðµ
    plt.subplot(2, 1, 1)
    plt.plot(prices, color='black', alpha=0.3, label='BTC/USD Close')

    for i, p in enumerate(preds):
        if p == 2: # BUY
            plt.scatter(i, prices[i], color='green', marker='^', s=100, edgecolors='black', label='BUY' if i==0 else "")
        elif p == 0: # SELL
            plt.scatter(i, prices[i], color='red', marker='v', s=100, edgecolors='black', label='SELL' if i==0 else "")

    plt.title("BTC/USD Trading Signals (Real Data, 3 Epochs)")
    plt.ylabel("Price")
    plt.legend(loc='best')
    plt.grid(True, alpha=0.2)

    # Ð“Ñ€Ð°Ñ„Ð¸Ðº 2: Ð›Ð°Ñ‚ÐµÐ½Ñ‚Ð½Ð¾Ðµ Ð¿Ñ€Ð¾ÑÑ‚Ñ€Ð°Ð½ÑÑ‚Ð²Ð¾ (Alpha Map)
    plt.subplot(2, 1, 2)
    # Ð˜ÑÐ¿Ð¾Ð»ÑŒÐ·ÑƒÐµÐ¼ t-SNE Ð´Ð»Ñ ÑÐ½Ð¸Ð¶ÐµÐ½Ð¸Ñ Ñ€Ð°Ð·Ð¼ÐµÑ€Ð½Ð¾ÑÑ‚Ð¸ Ð´Ð¾ 2D
    tsne = TSNE(n_components=2, perplexity=30, random_state=42, init='pca', learning_rate='auto')
    lat_2d = tsne.fit_transform(np.array(lats))

    # Ð£Ð²ÐµÑ€ÐµÐ½Ð½Ð¾ÑÑ‚ÑŒ Ð¼Ð¾Ð´ÐµÐ»Ð¸ â€” ÑÑ‚Ð¾ Ð²ÐµÑ€Ð¾ÑÑ‚Ð½Ð¾ÑÑ‚ÑŒ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ð¾Ð³Ð¾ ÐºÐ»Ð°ÑÑÐ°
    conf = [pr[preds[i]] for i, pr in enumerate(probs)]

    scatter = plt.scatter(lat_2d[:, 0], lat_2d[:, 1], c=conf, cmap='plasma', alpha=0.8, edgecolors='white', linewidth=0.5)
    plt.colorbar(scatter, label='Model Confidence P(class)')
    plt.title("Latent Alpha Map (Feature Clustering)")
    plt.xlabel("TSNE Dimension 1")
    plt.ylabel("TSNE Dimension 2")
    plt.grid(True, alpha=0.1)

    plt.tight_layout()
    plt.show()

# Ð—Ð°Ð¿ÑƒÑÐº Ð¸ÑÐ¿Ñ€Ð°Ð²Ð»ÐµÐ½Ð½Ð¾Ð¹ Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ð¸
if 'full_dataset' in locals():
    plot_real_results(num_points=300)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch

def precise_backtest(initial_balance=10000, commission=0.001):
    """
    Ð¡Ñ‚Ñ€Ð¾Ð³Ð¸Ð¹ Ð±ÑÐºÑ‚ÐµÑÑ‚ ÑÑ‚Ñ€Ð°Ñ‚ÐµÐ³Ð¸Ð¸ Reversal (Long/Short).

    Ð›Ð¾Ð³Ð¸ÐºÐ°:
    - Ð¡Ð¸Ð³Ð½Ð°Ð» BUY (2): Ð—Ð°ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Short (ÐµÑÐ»Ð¸ ÐµÑÑ‚ÑŒ), Ð¾Ñ‚ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Long Ð½Ð° Ð²ÑÐµ Ð´ÐµÐ¿Ð¾.
    - Ð¡Ð¸Ð³Ð½Ð°Ð» SELL (0): Ð—Ð°ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Long (ÐµÑÐ»Ð¸ ÐµÑÑ‚ÑŒ), Ð¾Ñ‚ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Short Ð½Ð° Ð²ÑÐµ Ð´ÐµÐ¿Ð¾.
    - Ð¡Ð¸Ð³Ð½Ð°Ð» HOLD (1): ÐÐ¸Ñ‡ÐµÐ³Ð¾ Ð½Ðµ Ð´ÐµÐ»Ð°ÐµÐ¼, ÑÐ¸Ð´Ð¸Ð¼ Ð² Ñ‚ÐµÐºÑƒÑ‰ÐµÐ¹ Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ð¸.
    """

    model.eval()

    # 1. Ð¡Ð±Ð¾Ñ€ Ð´Ð°Ð½Ð½Ñ‹Ñ… (ÑÐ¸Ð³Ð½Ð°Ð»Ñ‹ Ð¸ Ñ†ÐµÐ½Ñ‹)
    # Ð‘ÐµÑ€ÐµÐ¼ Ð¿Ð¾ÑÐ»ÐµÐ´Ð½Ð¸Ðµ 1000 ÑÐ²ÐµÑ‡ÐµÐ¹ Ð´Ð»Ñ Ñ€ÐµÐ¿Ñ€ÐµÐ·ÐµÐ½Ñ‚Ð°Ñ‚Ð¸Ð²Ð½Ð¾ÑÑ‚Ð¸ (Ð¸Ð»Ð¸ Ð¼ÐµÐ½ÑŒÑˆÐµ, ÐµÑÐ»Ð¸ Ð´Ð°Ñ‚Ð°ÑÐµÑ‚ Ð¼ÐµÐ½ÑŒÑˆÐµ)
    test_len = min(1000, len(full_dataset))
    start_idx = len(full_dataset) - test_len

    print(f"ðŸ’¼ ÐÐÐ§ÐÐ›Ðž Ð‘Ð­ÐšÐ¢Ð•Ð¡Ð¢Ð")
    print(f"   Ð”ÐµÐ¿Ð¾Ð·Ð¸Ñ‚: ${initial_balance}")
    print(f"   ÐšÐ¾Ð¼Ð¸ÑÑÐ¸Ñ: {commission*100}%")
    print(f"   ÐŸÐµÑ€Ð¸Ð¾Ð´: {test_len} ÑÐ²ÐµÑ‡ÐµÐ¹")
    print("-" * 40)

    prices = []
    signals = []

    # Ð˜Ð·Ð²Ð»ÐµÑ‡ÐµÐ½Ð¸Ðµ Ð¿Ñ€Ð¾Ð³Ð½Ð¾Ð·Ð¾Ð² (Ð±ÐµÐ· Ð³Ñ€Ð°Ð´Ð¸ÐµÐ½Ñ‚Ð¾Ð² Ð´Ð»Ñ ÑÐºÐ¾Ñ€Ð¾ÑÑ‚Ð¸)
    with torch.no_grad():
        for i in range(start_idx, len(full_dataset)):
            x_seq, _ = full_dataset[i]
            x_t = torch.from_numpy(x_seq).float().unsqueeze(0).to(device)

            # TCN -> FC
            pred = model(x_t).argmax(1).item()

            # Ð¦ÐµÐ½Ð°
            real_idx = len(df) - len(full_dataset) + i
            price = df['Close'].iloc[real_idx]

            prices.append(price)
            signals.append(pred)

    # 2. Ð¢Ð¾Ñ€Ð³Ð¾Ð²Ñ‹Ð¹ Ñ†Ð¸ÐºÐ»
    balance = initial_balance # ÐšÑÑˆ (USD)
    position = 0.0            # ÐŸÐ¾Ð·Ð¸Ñ†Ð¸Ñ Ð² BTC (Ð¼Ð¾Ð¶ÐµÑ‚ Ð±Ñ‹Ñ‚ÑŒ Ð¾Ñ‚Ñ€Ð¸Ñ†Ð°Ñ‚ÐµÐ»ÑŒÐ½Ð¾Ð¹ Ð´Ð»Ñ ÑˆÐ¾Ñ€Ñ‚Ð°)
    equity_curve = []         # Ð˜ÑÑ‚Ð¾Ñ€Ð¸Ñ Ð¸Ð·Ð¼ÐµÐ½ÐµÐ½Ð¸Ñ ÑÑ‡ÐµÑ‚Ð°

    trades_count = 0

    for i in range(len(prices)):
        price = prices[i]
        signal = signals[i]

        # Ð¢ÐµÐºÑƒÑ‰Ð°Ñ Ð¾Ñ†ÐµÐ½ÐºÐ° Ð¿Ð¾Ñ€Ñ‚Ñ„ÐµÐ»Ñ (Mark-to-Market)
        # Ð•ÑÐ»Ð¸ Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ñ > 0: balance + (btc * price)
        # Ð•ÑÐ»Ð¸ Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ñ < 0: balance + (ÑˆÐ¾Ñ€Ñ‚_Ð¼Ð°Ñ€Ð¶Ð°) + PnL ...
        # Ð£Ð¿Ñ€Ð¾Ñ‰ÐµÐ½Ð½Ð°Ñ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð´Ð»Ñ Ñ€ÐµÐ²ÐµÑ€ÑÐ°: Equity = Cash + PositionValue
        # Cash Ñ…Ñ€Ð°Ð½Ð¸Ñ‚ Ð·Ð°Ð»Ð¾Ð³ Ð¸ ÑÐ²Ð¾Ð±Ð¾Ð´Ð½Ñ‹Ðµ ÑÑ€ÐµÐ´ÑÑ‚Ð²Ð°.

        # Ð”Ð»Ñ Ð¿Ñ€Ð¾ÑÑ‚Ð¾Ñ‚Ñ‹ Ð¼Ð¾Ð´ÐµÐ»Ð¸Ñ€Ð¾Ð²Ð°Ð½Ð¸Ñ Ñ€ÐµÐ²ÐµÑ€ÑÐ° (Long/Short):
        # ÐœÑ‹ Ñ…Ñ€Ð°Ð½Ð¸Ð¼ Equity Ð² Ð¼Ð¾Ð¼ÐµÐ½Ñ‚Ðµ.
        # ÐŸÑ€Ð¸ Ð²Ñ…Ð¾Ð´Ðµ Ð² Long: btc = Equity / price.
        # ÐŸÑ€Ð¸ Ð²Ñ…Ð¾Ð´Ðµ Ð² Short: btc = -Equity / price.

        # Ð’ Ð½Ð°Ñ‡Ð°Ð»Ðµ ÑˆÐ°Ð³Ð° ÑÑ‡Ð¸Ñ‚Ð°ÐµÐ¼ Equity
        if i == 0:
            current_equity = balance
        else:
            # Ð˜Ð·Ð¼ÐµÐ½ÐµÐ½Ð¸Ðµ Ñ†ÐµÐ½Ñ‹
            price_change_pct = (price - prices[i-1]) / prices[i-1]

            # Ð•ÑÐ»Ð¸ Ð¼Ñ‹ Ð² Ð»Ð¾Ð½Ð³Ðµ, Ð½Ð°Ñˆ Equity Ñ€Ð°ÑÑ‚ÐµÑ‚ Ð½Ð° change_pct
            # Ð•ÑÐ»Ð¸ Ð¼Ñ‹ Ð² ÑˆÐ¾Ñ€Ñ‚Ðµ, Ð½Ð°Ñˆ Equity Ð¿Ð°Ð´Ð°ÐµÑ‚ Ð½Ð° change_pct
            # Ð•ÑÐ»Ð¸ Ð² ÐºÑÑˆÐµ (0), Ð½Ðµ Ð¼ÐµÐ½ÑÐµÑ‚ÑÑ

            # Ð¢Ð¾Ñ‡Ð½Ñ‹Ð¹ Ñ€Ð°ÑÑ‡ÐµÑ‚ Ñ‡ÐµÑ€ÐµÐ· Ð¿Ð¾Ð·Ð¸Ñ†Ð¸ÑŽ:
            # Equity = ÐŸÑ€ÐµÐ´Ñ‹Ð´ÑƒÑ‰ÐµÐµ_Equity * (1 + (Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ñ_Ð·Ð½Ð°Ðº * price_change_pct))
            # ÐÐ¾ ÑÑ‚Ð¾ Ð³Ñ€ÑƒÐ±Ð¾. Ð”Ð°Ð²Ð°Ð¹Ñ‚Ðµ Ñ‡ÐµÑ€ÐµÐ· "Ð²Ð¸Ñ€Ñ‚ÑƒÐ°Ð»ÑŒÐ½Ñ‹Ð¹ Ð±Ð°Ð»Ð°Ð½Ñ".
            pass

    # --- ÐŸÐ•Ð Ð•ÐŸÐ˜Ð¡ÐÐÐÐ«Ð™ Ð¦Ð˜ÐšÐ› (Ð§Ð•Ð¡Ð¢ÐÐ«Ð™ Ð£Ð§Ð•Ð¢) ---
    # Cash - ÑÑ‚Ð¾ ÑÐ²Ð¾Ð±Ð¾Ð´Ð½Ñ‹Ðµ Ð´Ð¾Ð»Ð»Ð°Ñ€Ñ‹.
    # BTC - ÑÑ‚Ð¾ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð°ÐºÑ‚Ð¸Ð²Ð°.
    # Short Ñ€ÐµÐ°Ð»Ð¸Ð·ÑƒÐµÐ¼ ÐºÐ°Ðº Ð¾Ñ‚Ñ€Ð¸Ñ†Ð°Ñ‚ÐµÐ»ÑŒÐ½Ñ‹Ð¹ BTC.

    # ÐÐ° ÑÑ‚Ð°Ñ€Ñ‚Ðµ Ñƒ Ð½Ð°Ñ Ñ‚Ð¾Ð»ÑŒÐºÐ¾ Cash
    cash = initial_balance
    btc_pos = 0.0
    equity_curve = [initial_balance]

    last_signal = 1 # Ð¡Ñ‡Ð¸Ñ‚Ð°ÐµÐ¼, Ñ‡Ñ‚Ð¾ Ð´Ð¾ Ð½Ð°Ñ‡Ð°Ð»Ð° Ð¼Ñ‹ Ð±Ñ‹Ð»Ð¸ Ð² HOLD (Cash)

    for i in range(len(prices)):
        price = prices[i]
        sig = signals[i]

        # 2.1 Ð˜ÑÐ¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ðµ ÑÐ¸Ð³Ð½Ð°Ð»Ð¾Ð²
        # BUY (2)
        if sig == 2:
            if btc_pos <= 0: # Ð•ÑÐ»Ð¸ Ð¼Ñ‹ Ð² ÐšÑÑˆÐµ Ð¸Ð»Ð¸ Ð¨Ð¾Ñ€Ñ‚Ðµ -> ÐŸÐžÐšÐ£ÐŸÐÐ•Ðœ
                # 1. Ð—Ð°ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Ð¨Ð¾Ñ€Ñ‚ (ÐµÑÐ»Ð¸ Ð±Ñ‹Ð»)
                if btc_pos < 0:
                    # ÐžÑ‚ÐºÑƒÐ¿ ÑˆÐ¾Ñ€Ñ‚Ð°: Ð¿Ð»Ð°Ñ‚Ð¸Ð¼ (Ñ†ÐµÐ½Ð° * ÐºÐ¾Ð»-Ð²Ð¾)
                    cost = abs(btc_pos) * price
                    fee = cost * commission
                    cash -= (cost + fee) # Ð¡Ð¿Ð¸ÑÑ‹Ð²Ð°ÐµÐ¼ Ð´ÐµÐ½ÑŒÐ³Ð¸ Ð·Ð° Ð¾Ñ‚ÐºÑƒÐ¿ + ÐºÐ¾Ð¼ÑÑƒ
                    # Ð¢ÐµÐ¿ÐµÑ€ÑŒ btc_pos = 0, cash = Ð²ÐµÑÑŒ ÐºÐ°Ð¿Ð¸Ñ‚Ð°Ð» (Equity)
                    btc_pos = 0

                # 2. ÐžÑ‚ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Ð›Ð¾Ð½Ð³ Ð½Ð° Ð²ÑÐµ
                # Ð”Ð¾ÑÑ‚ÑƒÐ¿Ð½Ð¾ Ð´ÐµÐ½ÐµÐ³: cash
                # ÐœÐ¾Ð¶ÐµÐ¼ ÐºÑƒÐ¿Ð¸Ñ‚ÑŒ: cash / price (Ñ ÑƒÑ‡ÐµÑ‚Ð¾Ð¼ ÐºÐ¾Ð¼Ð¸ÑÑÐ¸Ð¸)
                # cash = amount * price + amount * price * comm
                # cash = amount * price * (1 + comm)
                amount = cash / (price * (1 + commission))

                cash -= amount * price # Ð”ÐµÐ½ÑŒÐ³Ð¸ ÑƒÑˆÐ»Ð¸ Ð² Ð°ÐºÑ‚Ð¸Ð²
                # Ð¡Ð¿Ð¸ÑÑ‹Ð²Ð°ÐµÐ¼ ÐºÐ¾Ð¼Ð¸ÑÑÐ¸ÑŽ "Ð¸Ð· Ð²Ð¾Ð·Ð´ÑƒÑ…Ð°" (Ð¸Ð· Ð¾ÑÑ‚Ð°Ñ‚ÐºÐ° ÐºÑÑˆÐ°)? ÐÐµÑ‚, ÑÐ¿Ð¸ÑÑ‹Ð²Ð°ÐµÐ¼ ÐºÐ¾Ñ€Ñ€ÐµÐºÑ‚Ð½Ð¾.
                # Ð’Ñ‹ÑˆÐµ Ð¼Ñ‹ Ñ€Ð°ÑÑÑ‡Ð¸Ñ‚Ð°Ð»Ð¸ amount Ñ‚Ð°Ðº, Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ñ…Ð²Ð°Ñ‚Ð¸Ð»Ð¾ Ð¸ Ð½Ð° Ð¿Ð¾ÐºÑƒÐ¿ÐºÑƒ, Ð¸ Ð½Ð° ÐºÐ¾Ð¼Ð¸ÑÑÐ¸ÑŽ.
                cash -= amount * price * commission

                btc_pos = amount
                trades_count += 1

        # SELL (0)
        elif sig == 0:
            if btc_pos >= 0: # Ð•ÑÐ»Ð¸ Ð¼Ñ‹ Ð² ÐšÑÑˆÐµ Ð¸Ð»Ð¸ Ð›Ð¾Ð½Ð³Ðµ -> ÐŸÐ ÐžÐ”ÐÐ•Ðœ
                # 1. Ð—Ð°ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Ð›Ð¾Ð½Ð³ (ÐµÑÐ»Ð¸ Ð±Ñ‹Ð»)
                if btc_pos > 0:
                    revenue = btc_pos * price
                    fee = revenue * commission
                    cash += (revenue - fee)
                    btc_pos = 0

                # 2. ÐžÑ‚ÐºÑ€Ñ‹Ð²Ð°ÐµÐ¼ Ð¨Ð¾Ñ€Ñ‚ Ð½Ð° Ð²ÑÐµ
                # Ð’ "Ð´Ð¾Ð¼Ð°ÑˆÐ½ÐµÐ¼" Ð±ÑÐºÑ‚ÐµÑÑ‚Ðµ ÑˆÐ¾Ñ€Ñ‚ - ÑÑ‚Ð¾ Ð¿Ñ€Ð¾Ð´Ð°Ð¶Ð° Ñ‚Ð¾Ð³Ð¾, Ñ‡ÐµÐ³Ð¾ Ð½ÐµÑ‚, Ð¸ Ð¿Ð¾Ð»ÑƒÑ‡ÐµÐ½Ð¸Ðµ cash + Ð´Ð¾Ð»Ð³ Ð² BTC.
                # Equity = cash - abs(btc_pos)*price.
                # Ð§Ñ‚Ð¾Ð±Ñ‹ Ð¾Ñ‚ÐºÑ€Ñ‹Ñ‚ÑŒ ÑˆÐ¾Ñ€Ñ‚ Ð½Ð° Equity $1000 Ð¿Ñ€Ð¸ Ñ†ÐµÐ½Ðµ $100:
                # ÐœÑ‹ Ð¿Ñ€Ð¾Ð´Ð°ÐµÐ¼ 10 BTC. ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ $1000 ÐºÑÑˆÐ°. Ð˜Ñ‚Ð¾Ð³Ð¾ ÐºÑÑˆ $2000. Ð”Ð¾Ð»Ð³ 10 BTC.
                # Equity = 2000 - 10*100 = 1000.

                curr_equity = cash + btc_pos * price # (Ð·Ð´ÐµÑÑŒ btc_pos=0, Ñ‚Ð°Ðº Ñ‡Ñ‚Ð¾ Ð¿Ñ€Ð¾ÑÑ‚Ð¾ cash)

                # Ð¡ÐºÐ¾Ð»ÑŒÐºÐ¾ ÑˆÐ¾Ñ€Ñ‚Ð¸Ñ‚ÑŒ?
                # Ð¥Ð¾Ñ‚Ð¸Ð¼ ÑˆÐ¾Ñ€Ñ‚ Ð½Ð° Ð¾Ð±ÑŠÐµÐ¼ curr_equity.
                # amount = curr_equity / price
                # ÐŸÑ€Ð¸ Ð¿Ñ€Ð¾Ð´Ð°Ð¶Ðµ Ð¿Ð»Ð°Ñ‚Ð¸Ð¼ ÐºÐ¾Ð¼Ð¸ÑÑÐ¸ÑŽ.
                # Ð ÐµÐ°Ð»Ð¸Ð·ÑƒÐµÐ¼ ÑƒÐ¿Ñ€Ð¾Ñ‰ÐµÐ½Ð½Ð¾: "Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ñ = -amount"
                amount = curr_equity / (price * (1 + commission))

                # Ð­Ð¼ÑƒÐ»ÑÑ†Ð¸Ñ Ð¿Ñ€Ð¾Ð´Ð°Ð¶Ð¸ Ð² ÑˆÐ¾Ñ€Ñ‚:
                # ÐŸÐ¾Ð»ÑƒÑ‡Ð°ÐµÐ¼ ÐºÑÑˆ Ð¾Ñ‚ Ð¿Ñ€Ð¾Ð´Ð°Ð¶Ð¸
                cash += amount * price
                # ÐŸÐ»Ð°Ñ‚Ð¸Ð¼ ÐºÐ¾Ð¼Ð¸ÑÑÐ¸ÑŽ
                cash -= amount * price * commission

                btc_pos = -amount
                trades_count += 1

        # 2.2 Ð Ð°ÑÑ‡ÐµÑ‚ Equity Ð½Ð° ÐºÐ¾Ð½ÐµÑ† ÑÐ²ÐµÑ‡Ð¸
        # Equity = ÐšÑÑˆ + Ð¡Ñ‚Ð¾Ð¸Ð¼Ð¾ÑÑ‚ÑŒ Ð¿Ð¾Ð·Ð¸Ñ†Ð¸Ð¸
        current_eq = cash + (btc_pos * price)
        equity_curve.append(current_eq)

    # 3. Ð¡Ñ‚Ð°Ñ‚Ð¸ÑÑ‚Ð¸ÐºÐ°
    final_balance = equity_curve[-1]
    total_ret = (final_balance - initial_balance) / initial_balance * 100
    buy_hold_ret = (prices[-1] - prices[0]) / prices[0] * 100

    # 4. Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ
    plt.figure(figsize=(12, 6))
    plt.plot(equity_curve, label='AI Strategy Equity', color='blue', linewidth=2)

    # Ð¡Ñ€Ð°Ð²Ð½Ð¸Ð²Ð°ÐµÐ¼ Ñ Buy & Hold (Ð½Ð¾Ñ€Ð¼Ð¸Ñ€ÑƒÐµÐ¼ Ð¿Ð¾Ð´ ÑÑ‚Ð°Ñ€Ñ‚Ð¾Ð²Ñ‹Ð¹ Ð±Ð°Ð»Ð°Ð½Ñ)
    bh_curve = [initial_balance * (p / prices[0]) for p in prices]
    plt.plot(bh_curve, label='Buy & Hold BTC', color='gray', alpha=0.5, linestyle='--')

    plt.axhline(y=initial_balance, color='red', linestyle=':', alpha=0.5)
    plt.title(f"Backtest Result: {total_ret:.2f}% vs Buy&Hold {buy_hold_ret:.2f}%")
    plt.ylabel("Portfolio Value ($)")
    plt.xlabel("Ð¡Ð²ÐµÑ‡Ð¸")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    print("\n" + "="*30)
    print("ðŸ“Š Ð Ð•Ð—Ð£Ð›Ð¬Ð¢ÐÐ¢Ð« Ð‘Ð­ÐšÐ¢Ð•Ð¡Ð¢Ð")
    print("="*30)
    print(f"Ð¡Ñ‚Ð°Ñ€Ñ‚:       ${initial_balance:,.2f}")
    print(f"Ð¤Ð¸Ð½Ð¸Ñˆ:       ${final_balance:,.2f}")
    print(f"Ð§Ð¸ÑÑ‚Ð°Ñ Ð¿Ñ€Ð¸Ð±Ñ‹Ð»ÑŒ: ${final_balance - initial_balance:,.2f}")
    print(f"Ð”Ð¾Ñ…Ð¾Ð´Ð½Ð¾ÑÑ‚ÑŒ:  {total_ret:.2f}%")
    print(f"Ð¡Ð´ÐµÐ»Ð¾Ðº:      {trades_count}")
    print("-" * 30)
    if total_ret > buy_hold_ret:
        print("âœ… Ð¡Ñ‚Ñ€Ð°Ñ‚ÐµÐ³Ð¸Ñ Ð¾Ð±Ð¾Ð³Ð½Ð°Ð»Ð° Buy & Hold!")
    else:
        print("âŒ Ð¡Ñ‚Ñ€Ð°Ñ‚ÐµÐ³Ð¸Ñ Ð¿Ñ€Ð¾Ð¸Ð³Ñ€Ð°Ð»Ð° Ñ€Ñ‹Ð½ÐºÑƒ.")

if 'full_dataset' in locals():
    precise_backtest(10000)
```
